```python
import io, time, json
import requests as req
from bs4 import BeautifulSoup
import pandas as pd
from pandas.io.json import json_normalize
```

```python
class dataclean(object):
    
    def __init__(self,filename):
        with open(filename, 'r') as f:
            self.data=json.load(f)
    
    def extract(self,row,tag):
        df1=[]
        if tag=='restaurants':
            for i in range(len(row)):
                df1.append(row[i][0]['restaurant']['name'])
        if tag=='urls':
            for i in range(len(row)):
                 df1.append(row[i][0]['restaurant']['url'])
        return df1
            
    def dfr(self):
        #print(self.data[0]['restaurants'][0]['restaurant'])
        df=json_normalize(self.data)
        col=['results_shown','results_start','status','code','message','results_found']
        c=list(df.columns)
        c.remove('restaurants')
        df=df.drop(c ,axis=1)
        df=df.dropna()
        df=df[df.astype(str)['restaurants'] != '[]']
        df1=df.apply(self.extract,args=['restaurants'],axis=1)
        df1['urls']=df.apply(self.extract,args=['urls'],axis=1)
        return df1
```

```python
filename='file{}.json'
for i in range(1,6):
    if i==1:
        filenam=filename.format(i)
        dat=dataclean(filenam)
        dff=dat.dfr()
    if i>1:
        filenam=filename.format(i)
        dat=dataclean(filenam)
        dfn=dat.dfr()
        dff=dff.append(dfn,ignore_index=True)
dff
```
